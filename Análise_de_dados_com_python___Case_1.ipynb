# Case

Levantamento de dados para que um cliente encontre uma casa, conforme suas expectativas.

```python
import pandas as pd
```

```python
aluguel_casa = pd.read_excel('/content/houses_rent.xlsx') # importando a base excel usando pandas.

aluguel_casa.head() # head exibe as 05 primeiras linhas do dataset
```

```python
aluguel_casa.tail(20) # tail exibe as 05 últimas linhas do dataset -- para inserir mais de 05, só incluir em () a quantidade de linhas solicitadas
```

```python
aluguel_casa = aluguel_casa.rename(columns= {"parking spaces":"parking_spaces"}) # alterar o nome de colunas
aluguel_casa.head()
```

```python
aluguel_casa.dtypes # ver os dados e os seus tipos
```

```python
aluguel_casa.astype({"rent amount":"string"}) # alterar o tipo de dados
aluguel_casa.dtypes
```

```python
aluguel_casa.info() # informações mais detalhadas do que o dtype, incluindo a contagem de dados não-nulos, trazendo mais precisão a dados nulos no dataset
```

```python
aluguel_casa[["area","rooms","bathroom"]].describe() # exibe estatísticas descritivas sobre colunas numéricas em um dataframe, sendo:
# count - contagem
# média - média
# std - desvio padrão
# min - o menor valor apresentado
# 25% - limite superior do primeiro quartil (25% menores valores) | limite superior dos meus primeiros dados, estando entre o mínimo e 25%
# 50% - limite superior do segundo quartil (mediana)
# 75% - limite superior do terceiro quartil (75% menores valores)
# max - valor máximo
```

# Filtrar dados

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)] # filtrar os dados
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept")] # filtrar mais de um dado, usando o &
aluguel_casa.head()
```

```python
aluguel_casa.query('total <= 4000 and animal == "acepta"') # opção para filtrar com dados mais simples
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2)|(aluguel_casa["rooms"]==3))] # Para utilizar a opção ou, utilizamos |
aluguel_casa.head()
```

```python
aluguel_casa.describe() # dados modificados
```

# Agregação

```python
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean") # média de preço por cidade | Agg agrupa baseado nas informações solicitadas.
agg_city_price
```

```python
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
agg_bathroom_price = aluguel_casa.groupby(["bathroom"])["total"].agg("mean").reset_index()
agg_bathroom_price
```

# Biblioteca Seaborg

Visualização de dados

```python
import seaborn as sns

ax= sns.barplot(data=agg_bathroom_price, x="bathroom", y="total", color = "purple") # gráfico de barras
ax.set (xlabel = "Quantidade de banheiro", ylabel = "Preço", title = "Preço por Quantidade de Quartos")
```
